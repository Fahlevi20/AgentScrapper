# Load data from web

```python
import asyncio
from langchain_community.document_loaders import AsyncChromiumLoader


async def main():
    # Define the URLs
    urls = ['https://www.ebay.com/b/Dell-G5-15-PC-Laptops-Netbooks/177/bn_7116856782']

    # Create the loader
    loader = AsyncChromiumLoader(urls)

    # Load content asynchronously
    html = await loader.ascrape_playwright(urls[0])  # Call the asynchronous scrape method for one URL
    return html


# Adjust for compatibility with environments like Jupyter Notebook
try:
    html_result = asyncio.run(main())
except RuntimeError:  # Running in an event loop already
    html_result = asyncio.create_task(main())

print(html_result)
```
